## Year 2022/2023

```r
# Create dataframe with


library(ggplot2)


library(readxl)


library(dplyr)

library(openxlsx)


library(pdftools)
```

```r
### Importing dataset:

grape_dat = read_excel("G:\\My Drive\\Doringrivier\\Data\\Cultivar production_price.xlsx" ,"Sheet2" )

```

```r
head(grape_dat)
```

```r

# Let's get only the 2023 data:

grape_dat_2023 = as.data.frame(grape_dat[grape_dat$Year == 2023,])

print(grape_dat_2023)

ggplot(grape_dat, aes(x = factor(Cultivar), y = Income_hect, fill = Cultivar)) +
  geom_bar(stat = "identity")


my_colors <- c("Red", "Green", "Blue", "Black","Yellow","Pink")  # Replace with your desired colors

# Bar chart with custom colors
ggplot(data = grape_dat_2023, aes(x = Cultivar, y = Income_hect, fill = Cultivar)) +

geom_bar(stat = "identity") +

scale_fill_manual(values = my_colors) +

theme_minimal() +



 scale_y_continuous(labels = function(x) format(x, scientific = FALSE)) +
  labs(title = "Bar chart van inkomstes per hektaar vir elke kultivar 2023",
       x = "Cultivars",
       y = "Average income/hectar")
```
